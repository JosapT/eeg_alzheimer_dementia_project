```python
import pandas as pd
import os
import mne
import numpy as np
import scipy

DIR =  "/Users/josephthi/Desktop/UCI_Classes/cs184A/CS_184A_Final Project/ds004504/derivatives/"
```

```python
eeg_file_raw = r"/Users/josephthi/Desktop/UCI_Classes/cs184A/CS_184A_Final Project/ds004504/sub-001/eeg/sub-001_task-eyesclosed_eeg.set"
raw_eeg = mne.io.read_raw_eeglab(eeg_file_raw, preload=True)
print(raw_eeg.info)
raw_eeg.plot()
```

#### mne.io.read_raw_eeglab will return a tuple. First element is matrix of data. Second element is timestamp

```python
# Make sure channel names are all the same across all files
channels = raw_eeg.info.ch_names
print(channels)
for sub in os.listdir(DIR):
    if "sub-" in sub:
        tmp_eeg = mne.io.read_raw_eeglab(f"{DIR}/{sub}/eeg/{sub}_task-eyesclosed_eeg.set", preload=True)
        assert(channels == tmp_eeg.info.ch_names)
```

```python
# Downsample rate from 500hz to 256hz
# Hz is 128 so we can sample 64 hz freq. 
# https://nhahealth.com/brainwaves-the-language/
#    Gamme waves ~ 40 hz, if below, then possible learning disability


for sub in os.listdir(DIR):
    if "sub-" in sub:
        eeg = mne.io.read_raw_eeglab(f"{DIR}/{sub}/eeg/{sub}_task-eyesclosed_eeg.set", preload=True)
        eeg = eeg.resample(128)
        spectrogram = scipy.signal.spectrogram(eeg[:][0][1], fs=128)


        np.save(f"spectrogram/{sub}_x.npy", spectrogram[0]) #Frequencies
        np.save(f"spectrogram/{sub}_y.npy", spectrogram[1]) #times
        np.save(f"spectrogram/{sub}_spe.npy", spectrogram[2]) #image
        print(spectrogram[0].shape, spectrogram[1].shape)
        print(f"{sub} array saved")

        
```
